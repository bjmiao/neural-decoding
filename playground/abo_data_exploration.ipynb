##### Experiment Containers & Experiment
According to the official demo, the experiment container 'describes a set of experiments performed with the same targeted area, imaging depth, and Cre line'. There are 540 containers with 2534 experiments.

Unlike `ecephys_project_cache` data, in the ABO dataset where the neural activity is measured by 2-photon calcium imaging, each experiment can only measure `one place` of the brain. Therefore each experiments will have the following (fixed) parameters threefolds:
    1. `id`, `experiment_container_id`
    2. Stimuli protocol: `session_type`
    3. specimen and genotype information: `cre_line`, `reporter_line`, `donor_name`, `specimen_name`, `acquisition_age_days`
    4. the measured region: `imaging_depth`, `targeted_structures`
    5. additional information: `tags` and `fail_eye_tracking`

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def plot_stimulus_table(stim_table, title):
    fstart = stim_table.start.min()
    fend = stim_table.end.max()
    
    fig = plt.figure(figsize=(15,1))
    ax = fig.gca()
    for i, trial in stim_table.iterrows():    
        x1 = float(trial.start - fstart) / (fend - fstart)
        x2 = float(trial.end - fstart) / (fend - fstart)            
        ax.add_patch(patches.Rectangle((x1, 0.0), x2 - x1, 1.0, color='r'))
    ax.set_xticks((0,1))
    ax.set_xticklabels((fstart, fend))
    ax.set_yticks(())
    ax.set_title(title)
    ax.set_xlabel("frames")
```

```python
# the natural scenes stimulus table describes when each scene is on the screen
stim_table = exp.get_stimulus_table('natural_scenes')
print(len(stim_table), sorted(stim_table.frame.unique()))
print(stim_table.head())
# notice -1 means a black

# build up a mask of trials for which one of a list of scenes is visible
trial_mask = stim_table.frame == -2  ## initialize to False, since no -2 tag
demo_scene = [-1]
for scene in demo_scene:
    trial_mask |= (stim_table.frame == scene)
stim_table_show = stim_table[trial_mask]
plot_stimulus_table(stim_table_show, "scenes %s " % scene_nums)
```
